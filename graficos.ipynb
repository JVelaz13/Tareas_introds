```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Dannap7337/Pato/main/Base_limpiaclase.csv')
df=df.drop(columns=['Unnamed: 0'])
df
```

Gráficos de distribución (KDE Plot)

```python
#KDE para balance

plt.figure(figsize=(10,4))
sns.kdeplot(df["Balance"], fill = True, color = "#26B67E")
plt.title("Grafica de densidad de balance")
sns.set_style("whitegrid")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(df["Salario"], fill = True, color = "#2634B6")
plt.title("Grafica de densidad del salario")
sns.set_style("whitegrid")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(df["CreditScore"], fill = True, color = "#B626AA")
plt.title("Grafica de densidad de puntuacion de credito")
sns.set_style("whitegrid")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(df["Edad"], fill = True, color = "#B99037")
plt.title("Grafica de densidad de edad")
sns.set_style("whitegrid")
plt.show()
```

```python
# Crear figura y subplots (2 filas, 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(6, 4))

# KDE para Balance
sns.kdeplot(df['Balance'], fill=True, ax=axes[0, 0],color = "#5CB937")
axes[0, 0].set_title('Distribución de Balance')

# KDE para Salario
sns.kdeplot(df['Salario'], fill=True, ax=axes[0, 1],color = "#3774B9")
axes[0, 1].set_title('Distribución de Salario')

# KDE para CreditScore
sns.kdeplot(df['CreditScore'], fill=True, ax=axes[1, 0], color = "#B93737")
axes[1, 0].set_title('Distribución de CreditScore')

# KDE para Edad
sns.kdeplot(df['Edad'], fill=True, ax=axes[1, 1], color = "#AA8E10")
axes[1, 1].set_title('Distribución de Edad')

# Ajustar el layout para evitar superposición de títulos y etiquetas
plt.tight_layout()

# Mostrar gráfico
plt.show()
```

Gráficos de pastel

```python
# Contar la cantidad de clientes por género
gender_counts = df['Genero'].value_counts()

# Gráfico de pastel
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.2f%%', startangle=180,
         colors=['skyblue', 'lightgreen'])
plt.title('Distribución por Género')
plt.show()
```

```python
# Contar la cantidad de clientes por país
pais_counts = df['Pais'].value_counts()

# Gráfico de pastel
plt.figure(figsize=(6,6))
plt.pie(pais_counts, labels=pais_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'skyblue', 'lightgreen', 'yellow'])
plt.title('Distribución por País')
plt.show()
```

Gráficos de violín

```python
# Gráfico de violín para CreditScore por Tipo de Tarjeta
sns.violinplot(x='CreditScore', data=df, color = "#5DB370")

# Mostrar gráfico

plt.title('Distribución de CreditScore')
plt.show()
```
